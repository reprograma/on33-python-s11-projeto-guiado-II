```python
import sqlite3
```

```python
#NÃºmero de Estudantes que Participam de Atividades FÃ­sicas e o Tipo de Atividades
# conexÃ£o 
conn = sqlite3.connect(':memory:')
#escrever o df em uma tabela sql
df.to_sql('df', conn, index=False, if_exists='replace') 

#executar a consulta 
query_sql = """
SELECT envolvimento_esportes, COUNT(*) AS total_estudantes, GROUP_CONCAT(DISTINCT atividades_alivio_estresse) AS atividades
FROM df
GROUP BY envolvimento_esportes;
"""
crescimento_P = pd.read_sql_query(query_sql, conn)
print(crescimento_P )

#fechar a conexÃ£o
conn.close()
```

```python
# MÃ©dia de Ansiedade e DepressÃ£o por NÃ­vel de Curso


# conexÃ£o 
conn = sqlite3.connect(':memory:')
#escrever o df em uma tabela sql
df.to_sql('df', conn, index=False, if_exists='replace') 

#executar a consulta 
query_sql = """
SELECT nivel_do_curso, AVG(ansiedade) AS media_ansiedade, AVG(depression) AS media_depressao
FROM df
GROUP BY nivel_do_curso;
"""
crescimento_P = pd.read_sql_query(query_sql, conn)
print(crescimento_P )

#fechar a conexÃ£o
conn.close()
```

```python

from scipy import stats
import seaborn as sns

```

```python
# HÃ¡ uma diferenÃ§a significativa nos nÃ­veis de ansiedade entre estudantes que praticam atividades fÃ­sicas (envolvimento em esportes) e aqueles que nÃ£o praticam?

# Dividir os dados em dois grupos
com_esportes = df[df['envolvimento_esportes'] != 'No Sports']['ansiedade']
sem_esportes = df[df['envolvimento_esportes'] == 'No Sports']['ansiedade']

# Realizar o teste t
t_stat, p_value = stats.ttest_ind(com_esportes, sem_esportes)

# Exibir os resultados
print(f'EstatÃ­stica t: {t_stat}')
print(f'Valor p: {p_value}')

# Verificar se rejeitamos a hipÃ³tese nula
alpha = 0.05
if p_value < alpha:
    print("Rejeitamos a hipÃ³tese nula. HÃ¡ uma diferenÃ§a significativa na ansiedade entre estudantes que fazem atividades fÃ­sicas e aqueles que nÃ£o fazem.")
else:
    print("Falhamos em rejeitar a hipÃ³tese nula. NÃ£o hÃ¡ uma diferenÃ§a significativa na ansiedade entre estudantes que fazem atividades fÃ­sicas e aqueles que nÃ£o fazem.")


# Criar o grÃ¡fico de boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='envolvimento_esportes', y='ansiedade', data=df)
plt.title('DistribuiÃ§Ã£o da Ansiedade por Envolvimento em Esportes')
plt.xlabel('Envolvimento em Esportes')
plt.ylabel('Ansiedade')
#plt.xticks(rotation=45)  # Rotaciona os rÃ³tulos do eixo x para melhor visualizaÃ§Ã£o
plt.show()

# Mediana: A linha dentro de cada caixa representa a mediana dos dados, permitindo uma comparaÃ§Ã£o direta dos nÃ­veis de ansiedade entre os dois grupos.

# DispersÃ£o: A altura das caixas mostra a variaÃ§Ã£o dos dados dentro de cada grupo, indicando a consistÃªncia ou variabilidade dos nÃ­veis de ansiedade.

# Outliers: O boxplot tambÃ©m facilita a identificaÃ§Ã£o de outliers, que sÃ£o valores atÃ­picos que podem influenciar a anÃ¡lise.



    
```
