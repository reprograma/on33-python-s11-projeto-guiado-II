```python
# Parkinson e depressão
# Hipótese:
# H0: Não há correlação entre o diagnóstico de Parkinson e depressão.
# H1: Há correlação entre o diagnóstico de Parkinson e depressão.

from scipy.stats import ttest_ind

# Separando as amostras
depressao_com_parkinson = df[df['Diagnóstico_Parkinson'] == 1]['Depressão'].dropna()
depressao_sem_parkinson = df[df['Diagnóstico_Parkinson'] == 0]['Depressão'].dropna()

# Teste T
estatistica_t, valor_p = ttest_ind(depressao_com_parkinson, depressao_sem_parkinson)

# Resultados do teste
print("Teste T de depressão e diagnóstico de Parkinson")
print(f"Estatística T: {estatistica_t}")
print(f"Valor P: {valor_p}")

# Gráfico para visualização
sns.histplot(depressao_com_parkinson, color='blue', label='Com Parkinson', kde=True, bins=20)
sns.histplot(depressao_sem_parkinson, color='green', label='Sem Parkinson', kde=True, bins=20)

# Configurações do gráfico
plt.legend()
plt.title("Distribuição de Depressão em Pacientes com e sem Parkinson")
plt.xlabel("Depressão (0 = Não, 1 = Sim)")
plt.ylabel("Contagem")
plt.show()

# Interpretação do resultado
if valor_p < 0.05:
    print("Rejeitamos a hipótese nula: A depressão influencia o diagnóstico de Parkinson.")
else:
    print("Não rejeitamos a hipótese nula: A depressão não influencia o diagnóstico de Parkinson.")
```

```python
# Hipótese:
# H0: Não há correlação entre o valor de UPDRS do Parkinson e depressão.
# H1: Há correlação entre o valor de UPDRSdo Parkinson e depressão.

# Filtrar para pacientes com Parkinson (Diagnosis = 1)
df_parkinson = df[df['Diagnóstico_Parkinson'] == 1]

# Separar os pacientes em dois grupos: com depressão e sem depressão
updrs_com_depressao = df_parkinson[df_parkinson['Depressão'] == 1]['UPDRS']
updrs_sem_depressao = df_parkinson[df_parkinson['Depressão'] == 0]['UPDRS']

# Teste t para comparar as médias de UPDRS entre os dois grupos
estatistica_t, valor_p = ttest_ind(updrs_com_depressao, updrs_sem_depressao)

print("Teste T para UPDRS entre pacientes com e sem depressão")
print(f"Estatística T: {estatistica_t}")
print(f"Valor p: {valor_p}")

# Visualizar a distribuição de UPDRS em ambos os grupos
plt.figure(figsize=(8, 6))
sns.histplot(updrs_com_depressao, color='red', label='Com Depressão', kde=True, bins=20)
sns.histplot(updrs_sem_depressao, color='blue', label='Sem Depressão', kde=True, bins=20)
plt.title("Distribuição de UPDRS em Pacientes com e sem Depressão")
plt.xlabel("UPDRS Score")
plt.ylabel("Frequência")
plt.legend()
plt.show()

# Interpretação
if valor_p < 0.05:
    print("Rejeitamos a hipótese nula: A depressão está associada a uma diferença significativa na pontuação de UPDRS.")
else:
    print("Não rejeitamos a hipótese nula: Não há diferença significativa na pontuação de UPDRS entre pacientes com e sem depressão.")
```

```python
# Hipótese:
# H0: Não há correlação entre o valor de MoCA do Parkinson e depressão.
# H1: Há correlação entre o valor de MoCA do Parkinson e depressão.

# Separar os pacientes em dois grupos: com depressão e sem depressão
moca_com_depressao = df_parkinson[df_parkinson['Depressão'] == 1]['MoCA']
moca_sem_depressao = df_parkinson[df_parkinson['Depressão'] == 0]['MoCA']

# Teste t para comparar as médias de MoCA entre os dois grupos
estatistica_t_moca, valor_p_moca = ttest_ind(moca_com_depressao, moca_sem_depressao)

print("Teste T para MoCA entre pacientes com e sem depressão")
print(f"Estatística T: {estatistica_t_moca}")
print(f"Valor p: {valor_p_moca}")

# Visualizar a distribuição de MoCA em ambos os grupos
plt.figure(figsize=(8, 6))
sns.histplot(moca_com_depressao, color='red', label='Com Depressão', kde=True, bins=20)
sns.histplot(moca_sem_depressao, color='blue', label='Sem Depressão', kde=True, bins=20)
plt.title("Distribuição de MoCA em Pacientes com e sem Depressão")
plt.xlabel("Valor de MoCA")
plt.ylabel("Frequência")
plt.legend()
plt.show()

# Interpretação
if valor_p_moca < 0.05:
    print("Rejeitamos a hipótese nula: A depressão está associada a uma diferença significativa na pontuação de MoCA.")
else:
    print("Não rejeitamos a hipótese nula: Não há diferença significativa na pontuação de MoCA entre pacientes com e sem depressão.")
```

```python
# Hipótese:
# H0: Não há correlação entre o valor de Avaliação_Funcional do Parkinson e depressão.
# H1: Há correlação entre o valor de Avaliação_Funcional do Parkinson e depressão.

# Separar os pacientes em dois grupos: com depressão e sem depressão
func_assess_com_depressao = df_parkinson[df_parkinson['Depressão'] == 1]['Avaliação_Funcional']
func_assess_sem_depressao = df_parkinson[df_parkinson['Depressão'] == 0]['Avaliação_Funcional']

# Teste t para comparar as médias de Functional Assessment entre os dois grupos
estatistica_t_func_assess, valor_p_func_assess = ttest_ind(func_assess_com_depressao, func_assess_sem_depressao)

print("Teste T para Avaliação_Funcional entre pacientes com e sem depressão")
print(f"Estatística T: {estatistica_t_func_assess}")
print(f"Valor p: {valor_p_func_assess}")

# Visualizar a distribuição de Avaliação_Funcional em ambos os grupos
plt.figure(figsize=(8, 6))
sns.histplot(func_assess_com_depressao, color='red', label='Com Depressão', kde=True, bins=20)
sns.histplot(func_assess_sem_depressao, color='blue', label='Sem Depressão', kde=True, bins=20)
plt.title("Distribuição de Avaliação_Funcional em Pacientes com e sem Depressão")
plt.xlabel("Valor de Avaliação_Funcional")
plt.ylabel("Frequência")
plt.legend()
plt.show()

# Interpretação
if valor_p_func_assess < 0.05:
    print("Rejeitamos a hipótese nula: A depressão está associada a uma diferença significativa na pontuação de Functional Assessment.")
else:
    print("Não rejeitamos a hipótese nula: Não há diferença significativa na pontuação de Functional Assessment entre pacientes com e sem depressão.")
```

```python
# Depressão e a presença de sintomas do Parkinson

# Lista dos sintomas
sintomas = ['Tremor', 'Rigidez', 'Bradicinesia', 'Instabilidade_Postural', 'Problemas_de_Fala', 'Distúrbios_do_Sono', 'Constipação']

# Lista para armazenar as proporções de sintomas
dados_sintomas = []

# Calcular as proporções
for sintoma in sintomas:
    proporcao_com_depressao = df_parkinson[df_parkinson['Depressão'] == 1][sintoma].mean()
    proporcao_sem_depressao = df_parkinson[df_parkinson['Depressão'] == 0][sintoma].mean()
    
    dados_sintomas.append({
        'Sintoma': sintoma,
        'Com Depressão': proporcao_com_depressao * 100,
        'Sem Depressão': proporcao_sem_depressao * 100
    })

# Converter a lista para um DataFrame
proporcoes_sintomas = pd.DataFrame(dados_sintomas)

# Exibir as proporções
print(proporcoes_sintomas)

# Plotar as proporções em um gráfico de barras
proporcoes_sintomas.set_index('Sintoma').plot(kind='bar', figsize=(10, 6), color=['violet', 'skyblue'])
plt.title('Proporção de Sintomas em Pacientes com e sem Depressão')
plt.ylabel('Proporção (%)')
plt.xlabel('Sintoma')
plt.xticks(rotation=45)
plt.legend(['Com Depressão', 'Sem Depressão'])
plt.show()
```

```python
# Há correlação entre depressão e a presença de sintomas?

from scipy.stats import ttest_ind

# Lista dos sintomas
sintomas = [
    'Tremor', 
    'Rigidez', 
    'Instabilidade_Postural', 
    'Problemas_de_Fala', 
    'Distúrbios_do_Sono'
]

# Hipótese nula (H0): Não há correlação entre depressão e o sintoma.
# Hipótese alternativa (H1): Há correlação entre depressão e o sintoma.

# Loop para realizar o teste T para cada sintoma
for sintoma in sintomas:
    # Separar os dados em grupos: com e sem depressão
    com_depressao = df_parkinson[df_parkinson['Depressão'] == 1][sintoma].dropna()
    sem_depressao = df_parkinson[df_parkinson['Depressão'] == 0][sintoma].dropna()
    
    # Realizar o teste T
    estatistica_t, valor_p = ttest_ind(com_depressao, sem_depressao)
    
    # Exibir os resultados
    print(f"Teste T para {sintoma}:")
    print(f"Estatística T: {estatistica_t}")
    print(f"Valor P: {valor_p}")
    
    # Interpretar o resultado
    if valor_p < 0.05:
        print("Rejeitamos a hipótese nula: Há uma correlação entre depressão e o sintoma.\n")
    else:
        print("Não rejeitamos a hipótese nula: Não há evidências de correlação entre depressão e o sintoma.\n")
```
