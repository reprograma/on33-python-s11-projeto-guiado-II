# Análises

## Média do índice de felicidade - Top 20

```python
# Média do índice de felicidade (Life Ladder) para cada país
media_felicidade_paises = df.groupby('Nome do País')['Índice de Felicidade'].mean()

# Deixar em ordem decrescente e selecionar os 20 mais felizes
media_felicidade_paises_decresc = media_felicidade_paises.sort_values(ascending=False)
media_felicidade_paises_top20 = media_felicidade_paises.sort_values(ascending=False).head(20)

# Gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x=media_felicidade_paises_top20.values, y=media_felicidade_paises_top20.index, hue=media_felicidade_paises_top20.index, palette='coolwarm', legend=False)

plt.title('Os 20 Países Mais Felizes do Mundo', fontsize=16)
plt.xlabel('Média do Índice de Felicidade (Life Ladder)')
plt.ylabel('País')

plt.show()
```

### A posição do Brasil no ranking

```python
# Posição do Brasil - média
posicao_brasil = media_felicidade_paises_decresc.index.get_loc('Brazil') + 1  
# +1 para obter a posição na lista iniciando em 1

# Filtrar os dados para o ano de 2020
dados_2020 = df[df['Ano'] == 2020]

# Ordenar os países pelo índice de felicidade em ordem decrescente
paises_ordenados_2020 = dados_2020.sort_values(by='Índice de Felicidade', ascending=False)

# Encontrar a posição do Brasil
posicao_brasil_2020 = paises_ordenados_2020['Nome do País'].tolist().index('Brazil') + 1

print(f"O Brasil está na posição {posicao_brasil} no índice de felicidade.")
print(f"O Brasil está na posição {posicao_brasil_2020} no índice de felicidade em 2020.")
```

### Índice de felicidade no Brasil ao longo dos anos

```python
dados_brasil = df[df['Nome do País'] == 'Brazil']

plt.figure(figsize=(12, 6))
sns.lineplot(x='Ano', y='Índice de Felicidade', data=dados_brasil, marker='o')

plt.title('Evolução do Índice de Felicidade no Brasil ao Longo dos Anos', fontsize=16)
plt.xlabel('Ano')
plt.ylabel('Índice de Felicidade')
plt.show()
```

```python
# Plotar o Índice de Felicidade ao longo dos anos
plt.figure(figsize=(14, 8))

# Índice de Felicidade
plt.subplot(3, 2, 1)
sns.lineplot(x='Ano', y='Índice de Felicidade', data=dados_brasil, marker='o')
plt.title('Evolução do Índice de Felicidade no Brasil')
plt.xlabel('Ano')
plt.ylabel('Índice de Felicidade')

# PIB per capita (logaritmo)
plt.subplot(3, 2, 2)
sns.lineplot(x='Ano', y='PIB per capita (logaritmo)', data=dados_brasil, marker='o')
plt.title('Evolução do PIB per capita (logaritmo) no Brasil')
plt.xlabel('Ano')
plt.ylabel('PIB per capita (logaritmo)')

# Apoio Social
plt.subplot(3, 2, 3)
sns.lineplot(x='Ano', y='Apoio Social', data=dados_brasil, marker='o')
plt.title('Evolução do Apoio Social no Brasil')
plt.xlabel('Ano')
plt.ylabel('Apoio Social')

# Expectativa de vida saudável ao nascer
plt.subplot(3, 2, 4)
sns.lineplot(x='Ano', y='Expectativa de vida saudável ao nascer', data=dados_brasil, marker='o')
plt.title('Evolução da Expectativa de Vida Saudável no Brasil')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida Saudável')

# Liberdade para fazer escolhas
plt.subplot(3, 2, 5)
sns.lineplot(x='Ano', y='Liberdade para fazer escolhas', data=dados_brasil, marker='o')
plt.title('Evolução da Liberdade para Fazer Escolhas no Brasil')
plt.xlabel('Ano')
plt.ylabel('Liberdade para Fazer Escolhas')

# Percepções de corrupção
plt.subplot(3, 2, 6)
sns.lineplot(x='Ano', y='Percepções de corrupção', data=dados_brasil, marker='o')
plt.title('Evolução das Percepções de Corrupção no Brasil')
plt.xlabel('Ano')
plt.ylabel('Percepções de Corrupção')

plt.tight_layout()
plt.show()
```

### Índice de generosidade - Brasil vs Top 20

```python
dados_comparacao = df[df['Nome do País'].isin(media_felicidade_paises_top20.index.tolist() + ['Brazil'])]

media_generosidade = dados_comparacao.groupby('Nome do País')['Generosidade'].mean()

media_generosidade = media_generosidade.reset_index()

# Destacar o Brasil no gráfico
media_generosidade['Destaque'] = media_generosidade['Nome do País'].apply(lambda x: 'Brasil' if x == 'Brazil' else 'Outros')


plt.figure(figsize=(12, 8))
sns.barplot(x='Generosidade', y='Nome do País', data=media_generosidade, hue='Destaque', dodge=False, palette={'Brasil': 'blue', 'Outros': 'gray'})

plt.title('Comparação da Pontuação de Generosidade: Brasil vs. 20 Países Mais Felizes', fontsize=16)
plt.xlabel('Pontuação Média de Generosidade')
plt.ylabel('País')

plt.show()
```

# SQL

```python
import sqlite3
import pandas as pd

conexao = sqlite3.connect('felicidade_mundial.db')
cursor = conexao.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS tabela_felicidade (
    id INTEGER PRIMARY KEY,
    "Nome do País" TEXT,
    "Ano" INTEGER,
    "Índice de Felicidade" REAL,
    "PIB per capita (logaritmo)" REAL,
    "Apoio Social" REAL,
    "Expectativa de vida saudável ao nascer" REAL,
    "Liberdade para fazer escolhas" REAL,
    "Generosidade" REAL,
    "Percepções de corrupção" REAL,
    "Afeto positivo" REAL,
    "Afeto negativo" REAL
)
''')

df = pd.read_csv('world-happiness-report.csv')

df.to_sql('tabela_felicidade', conexao, if_exists='replace', index=False)
conexao.commit()
```

```python
# 20 países mais felizes de 2020

consulta_20_mais_felizes = '''
SELECT "Nome do País", "Índice de Felicidade"
FROM tabela_felicidade
WHERE "Ano" = 2020
ORDER BY "Índice de Felicidade" DESC
LIMIT 20;
'''
top_20_felizes = pd.read_sql_query(consulta_20_mais_felizes, conexao)
print("Os 20 países mais felizes de 2020:")
print(top_20_felizes)

# posição do Brasil em 2020
consulta_posicao_brasil = '''
SELECT "Nome do País", "Índice de Felicidade"
FROM tabela_felicidade
WHERE "Ano" = 2020
ORDER BY "Índice de Felicidade" DESC;
'''
todos_paises_2020 = pd.read_sql_query(consulta_posicao_brasil, conexao)

# posição do Brasil
posicao_brasil_2020 = todos_paises_2020[todos_paises_2020['Nome do País'] == 'Brazil'].index[0] + 1
print(f"O Brasil está na posição {posicao_brasil_2020} no índice de felicidade em 2020.")

conexao.close()
```
