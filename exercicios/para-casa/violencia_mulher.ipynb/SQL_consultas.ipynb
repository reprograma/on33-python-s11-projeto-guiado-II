```python
import sqlite3
import csv
import pandas as pd
```

```python
df_tratado = pd.read_csv('woman_tratado.csv')
```

```python
# conexão 
conn = sqlite3.connect(':memory:')

#escrever o df em uma tabela sql
df_tratado.to_sql('df_tratado', conn, index=False, if_exists='replace') 

#executar a consulta 
query_sql = """
SELECT MAx(Ano) as ano_maximo from df_tratado
"""

ano_maximo_df = pd.read_sql_query(query_sql, conn)
ano_maximo = ano_maximo_df['ano_maximo'].iloc[0]
print(f'O ano com maior número de registros de violência é', ano_maximo)

#fechar a conexão
conn.close()
```

```python
#conexão
conn = sqlite3.connect(':memory:')

df_tratado.to_sql('df_tratado', conn, index=False, if_exists='replace')

#executar a consulta
query_sql = """
SELECT COUNT(*) AS total_brasil
FROM df_tratado
WHERE pais = "Brazil";
"""

result_df = pd.read_sql_query(query_sql, conn)
total_brasil = result_df.at[0, 'total_brasil']
print(f"O número de registros de violência contra a mulher no Brasil é {total_brasil}.")

#fechar a conexão
conn.close()
```

```python
# conexão 
conn = sqlite3.connect(':memory:')

#escrever o df em uma tabela sql
df_tratado.to_sql('df_tratado', conn, index=False, if_exists='replace') 

#executar a consulta 
query_sql = """
SELECT "Medicao Nivel 1" AS medicao, COUNT(*) AS ocorrencias
FROM df_tratado
GROUP BY "Medicao Nivel 1"
ORDER BY ocorrencias DESC
LIMIT 3;
"""

result_df = pd.read_sql_query(query_sql, conn)
print("As principais formas de medir o tipo de violência, a nível geral, são:")
print(result_df)

#fechar a conexão
conn.close()
```
