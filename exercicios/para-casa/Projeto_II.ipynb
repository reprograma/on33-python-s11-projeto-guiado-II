Contar dados nulos por linha no DataFrame.

Seguindo, contar dados nulos por linha em um DataFrame, referentes aos anos de 2017, 2018 e 2019.
Obs: df.isnull() cria um DataFrame de valores booleanos onde True indica um valor nulo, ou seja: 0 para ausência, 1 para presença.
.sum(axis=1) soma os valores True (que representam os nulos) ao longo de cada linha. 
O argumento axis=1 especifica que a soma deve ser feita horizontalmente, ou seja, por linha.

```python
nulos_por_linha_2017 = df_2017.isnull().sum(axis=1)
nulos_por_linha_2018 = df_2018.isnull().sum(axis=1)
nulos_por_linha_2019 = df_2019.isnull().sum(axis=1)
```

```python
print("Dados nulos por linha em 2017:")
print(nulos_por_linha_2017)

print("\nDados nulos por linha em 2018:")
print(nulos_por_linha_2018)

print("\nDados nulos por linha em 2019:")
print(nulos_por_linha_2019)
```

O df.isnull() é utilizado para retornar um DataFrame do mesmo tamanho, com valores booleanos, onde True indica que o valor é nulo.
Já o df.isnull().sum() é utilizado para somar os valores True para cada coluna, resultando no número de valores nulos por coluna.
Abaixo, segue o número de valores nulos para cada coluna em cada um dos DataFrames de 2017, 2018, e 2019. Então o valor 0 impressos nas colunas significa a ausência de valores nulos e 1 a presença

```python
nulos_por_coluna_2017 = df_2017.isnull().sum()
nulos_por_coluna_2018 = df_2018.isnull().sum()
nulos_por_coluna_2019 = df_2019.isnull().sum()
```

Teste de hipótesse.

H1 = O rank de felicidade pode ter relação com o nível de economia do país.
H0 = Não existe relação entre o rank de felicidade e o nível de economia do país.

Para testar a hipótese de que o rank de felicidade tem relação com o nível de economia do país, foi utilizado o coeficiente de correlação de Pearson.

```python
from scipy.stats import pearsonr

corr_coef, p_value = pearsonr(df_2017['Economy..GDP.per.Capita.'], df_2017['Happiness.Score'])

if p_value < 0.05:
   print("Rejeitamos a hipótese nula. Há uma correlação significativa entre renda per capita e felicidade.")
   if corr_coef > 0:
     print("A correlação é positiva, indicando que países com maior renda per capita tendem a ser mais felizes.")
   else:
     print("A correlação é negativa, indicando que países com maior renda per capita tendem a ser menos felizes.")
else:
   print("Não rejeitamos a hipótese nula. Não há evidências suficientes para concluir que existe uma correlação entre renda per capita e felicidade.")
```

O coeficiente de correlação de Pearson é um teste que mede a relação estatística entre duas variáveis contínuas. Se a associação entre os elementos não for linear, o coeficiente não será representado adequadamente. O coeficiente de correlação de Pearson pode ter um intervalo de valores de +1 a -1.

```python
import pandas as pd
from scipy.stats import pearsonr

df_2017 = pd.read_csv("2017.csv")

renda = df_2017['Economy..GDP.per.Capita.']
felicidade = df_2017['Happiness.Score']

corr_coef, p_value = pearsonr(renda, felicidade)

print(f"Coeficiente de correlação: {corr_coef:.2f}")
print(f"Valor-p: {p_value:.3f}")

if p_value < 0.05:
  print("Rejeitamos a hipótese nula. Há uma correlação significativa entre renda per capita e felicidade.")
  if corr_coef > 0:
    print("A correlação é positiva, indicando que países com maior renda per capita tendem a ser mais felizes.")
  else:
    print("A correlação é negativa, indicando que países com maior renda per capita tendem a ser menos felizes.")
else:
  print("Não rejeitamos a hipótese nula. Não há evidências suficientes para concluir que existe uma correlação entre renda per capita e felicidade.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Para calcula a reta de regressão
x, y = np.polyfit(df_2017['Economy..GDP.per.Capita.'], df_2017['Happiness.Score'], 1)

# Para Criar o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(df_2017['Economy..GDP.per.Capita.'], df_2017['Happiness.Score'])

# Plotando a reta de regressão
plt.plot(df_2017['Economy..GDP.per.Capita.'], x * df_2017['Economy..GDP.per.Capita.'] + y, color='red')

# Configurações do gráfico
plt.xlabel('Renda Per Capita')
plt.ylabel('Pontuação de Felicidade')
plt.title('Relação entre Renda Per Capita e Felicidade')
plt.show()
```

```python
query = """
 SELECT Country, `Health..Life.Expectancy.`
 FROM df_2017
 ORDER BY `Health..Life.Expectancy.` DESC
 LIMIT 10;
 """
```
