```python
#Apagar colunas do DF
df = df.drop(columns=['English name', 'Japanese name', 'Aired', 'Premiered', 'Producers', 'Licensors', 'Studios', 'Source', 'Ranked', 'Popularity', 'Members','Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6','Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1'])

df.shape
```

```python
# renomear as colunas
df.rename(columns={
   'MAL_ID': 'AnimeId',
   'Name': 'Nome',
   'Score' : 'Pontuação',
   'Genres' : 'Genero',
   'Type' : 'Tipo',
   'Episodes' : 'Episodios',
   'Duration' : 'Duração',
   'Rating' : 'Faixa Etaria',
   'Favorites' : 'Favoritos',
   'Watching' : 'Assistidos',
   'Completed' : 'Completo',
   'On-Hold' : 'Em espera',
   'Dropped' : 'Excluidos',
   'Plan to Watch' : 'Planejam Assistir'
}, inplace=True)

df.columns
```

```python
#salvar no CSV
df.to_csv('Anime.Tratado.csv', index=False)
```

```python
# Substituir 'Unknown' por 0
df['Episodios'] = df['Episodios'].replace('Unknown', 0)

# Substituir valores vazios ou nulos por 0
df['Episodios'] = df['Episodios'].fillna(0).replace('', 0)

# Se a coluna não for numérica, você pode tentar convertê-la:
df['Episodios'] = pd.to_numeric(df['Episodios'], errors='coerce')
```

```python
#Modificando a coluna diração
df['Duração'] = df['Duração'].fillna('0 min. per ep.')  # Substituir None por um valor padrão
df['Duração'] = df['Duração'].str.replace(' min. per ep.', '')  # Remover o texto ' min. per ep.'
df['Duração'] = pd.to_numeric(df['Duração'], errors='coerce')  # Converter para numérico

print(df['Duração'].sum())

#OBS: Essa linha precisa ser executada uma de cada vez, senão vai dar erro :D
```

```python
#Convertendo a coluna pontuação pra float
df['Pontuação'] = pd.to_numeric(df['Pontuação'], errors='coerce')
```

```python
#Criando um sample pra utilizar em determinados graficos pq o data set é muito grande, para isso eu vou usar SQL


import sqlite3


# anime_amostra é o sample amostral do código
anime_amostra = df.sample(350)  # Amostra de 350 linhas do DataFrame

# Criar uma conexão com um banco de dados em memória
conn = sqlite3.connect(':memory:')

# Salvar o DataFrame no banco de dados SQLite
anime_amostra.to_sql('anime_amostra', conn, index=False, if_exists='replace') 

# Consultar dados usando SQL
query_sql = """
SELECT Tipo, COUNT(*) AS Contagem
FROM anime_amostra
GROUP BY Tipo;
"""

# Executar a consulta e armazenar o resultado em um DataFrame
contagem_por_tipo = pd.read_sql_query(query_sql, conn)
print(contagem_por_tipo)

# Fechar a conexão
conn.close()
```

## Testes Estatiscos

```python
# Calcular a popularidade como a soma das colunas especificadas
df['Popularidade'] = df[['Favoritos', 'Assistidos', 'Completo', 'Em espera', 'Excluidos', 'Planejam Assistir']].sum(axis=1)

# Mostrar o DataFrame atualizado
print(df['Popularidade'].head(10))
```

```python
from scipy.stats import ttest_ind

# Filtrar os dados para dois tipos de anime
tipo1 = df[df['Tipo'] == 'Tipo1']['Popularidade']
tipo2 = df[df['Tipo'] == 'Tipo2']['Popularidade']

# Realizar o teste T
t_stat, p_value = ttest_ind(tipo1, tipo2)

print(f'Testatística T: {t_stat}')
print(f'Valor-p: {p_value}')

# Interpretar o valor-p
alpha = 0.05
if p_value < alpha:
    print("O tipo de anime interfere na popularidade")
else:
    print("O tipo de anime não interfere na popularidade")
```

```python
# Teste Qui-Quardado

from scipy.stats import chi2_contingency

# tabela de contingência para 'Type' e 'Watching'
contingency_table = pd.crosstab(df['Tipo'], df['Assistidos'] > 0)

# Realizar o Teste Qui-Quadrado
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Tabela de Contingência:\n{contingency_table}\n")
print(f"Teste Qui-Quadrado: Chi2-stat = {chi2_stat:.2f}, p-value = {p_val:.2e}")
print(f"Graus de Liberdade: {dof}")
print(f"Valores Esperados:\n{expected}")

# Interpretação
alpha = 0.05 # Nível de significância
if p_val < alpha:
  print("\nResultado: Existe uma associação estatisticamente significativa entre o tipo de anime e a ação dos usuários.")
else:
  print("\nResultado: Não existe uma associação estatisticamente significativa entre o tipo de anime e a ação dos usuários.")
```
